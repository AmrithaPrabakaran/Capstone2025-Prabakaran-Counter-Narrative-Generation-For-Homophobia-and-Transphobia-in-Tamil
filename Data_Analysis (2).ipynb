**CATEGORY DISTRIBUTION**

```python
#Filtering unwanted categories
n_categories = ['Transphobic-derogation', 'Homophobic-derogation', 'Transphobic-Threatening', 'Homophobic-Threatening', 'Counter-speech']

# Filter each DataFrame
Cn_test_ndup = Cn_test_ndup[Cn_test_ndup['category'].isin(n_categories)]
Cn_train_ndup = Cn_train_ndup[Cn_train_ndup['category'].isin(n_categories)]
Cn_dev_ndup = Cn_dev_ndup[Cn_dev_ndup['category'].isin(n_categories)]
```

```python
Cn_test_ndup.to_excel('Cn_test_filtered.xlsx', index=False)
Cn_train_ndup.to_excel('Cn_train_filtered.xlsx', index=False)
Cn_dev_ndup.to_excel('Cn_dev_filtered.xlsx', index=False)
```

```python
sns.countplot(x='category', data=Cn_dev_ndup, color='red')
plt.xlabel("Category")
plt.ylabel("Count")
plt.title("Category Distribution After filtering in DEV dataset")
plt.xticks(rotation=45)
plt.show()
```
