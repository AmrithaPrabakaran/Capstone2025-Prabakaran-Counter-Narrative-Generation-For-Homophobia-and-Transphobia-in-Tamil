```python
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt


words = tamil_text.split()


stopwords = {'இது', 'தான்', 'என்று', 'அது', 'நான்', 'என', 'எனக்கு', 'அந்த', 'இருந்து'}
words = [w for w in words if w not in stopwords]


word_freq = Counter(words)


important_words = {
    'ஓரினச்சேர்க்கை': 500,
    'திருநங்கை': 400,
    'ஆண்': 350,
    'பெண்': 350
}
word_freq.update(important_words)  # This overrides with higher value

# Step 4: Use correct Tamil font with full glyph support
font_path = "/content/sample_data/NotoSansTamil-VariableFont_wdth,wght.ttf"  # or "Latha.ttf"

# Step 5: Generate word cloud
wordcloud = WordCloud(
    font_path=font_path,
    background_color='white',
    width=1200,
    height=600,
    max_words=300,
    min_font_size=8,
    scale=4,
    relative_scaling=0.6,
    collocations=False,
    prefer_horizontal=0.9
).generate_from_frequencies(word_freq)

# Step 6: Display
plt.figure(figsize=(15, 7))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

**SENTENCE LENGTH DISTRIBUTION**

```python

df = pd.read_excel("/content/Combined_Tamil_Text.xlsx")

text_col = 'text'
category_col = 'category'


categories = df[category_col].unique()
category_lengths = {}

for cat in categories:
    sentences = df[df[category_col] == cat][text_col].dropna().astype(str)
    lengths = [len(s.split()) for s in sentences]
    category_lengths[cat] = lengths


plt.figure(figsize=(10, 6))
colors = ['blue', 'red', 'green', 'orange']

for i, (cat, lengths) in enumerate(category_lengths.items()):
    plt.hist(lengths, bins=30, alpha=0.6, label=cat, color=colors[i % len(colors)])

plt.xlabel("Sentence Length (in words)", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.title("Sentence Length Distribution by Category", fontsize=14)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Evaluate Metrics for Existing Counter Narratives Dataset
import subprocess
import sys

# Ensure required packages are installed
required_packages = [
    "evaluate", "rouge_score", "transformers", "nltk", "torch", "pandas", "numpy", "bert_score"
]

def install(package):
    subprocess.check_call([sys.executable, "-m", "pip", "install", package])

for pkg in required_packages:
    try:
        __import__(pkg if pkg != "rouge_score" else "rouge_score")
    except ImportError:
        install(pkg)

from evaluate import load
import pandas as pd
import numpy as np
import torch
import nltk
from nltk.translate.bleu_score import sentence_bleu
from transformers import AutoTokenizer, AutoModelForCausalLM

nltk.download('punkt')

# Load dataset
file_path = "/content/sample_data/Final_dataset_Counter narratives_Tamil - Sheet1.csv"
df = pd.read_csv(file_path)
texts = df["TEXT"].dropna().astype(str).tolist()
cns = df["COUNTER NARRATIVES"].dropna().astype(str).tolist()

# BLEU, ROUGE, BERTScore
bleu = load("bleu")
rouge = load("rouge")
bertscore = load("bertscore")

# BLEU (unigram, bigram, trigram)
def compute_bleu(cns, n):
    weights = {1: (1.0,), 2: (0.5, 0.5), 3: (0.33, 0.33, 0.33)}[n]
    return np.mean([sentence_bleu([t.split()], p.split(), weights=weights) for t, p in zip(texts, cns)])

# Self-BLEU
def compute_self_bleu(cns):
    scores = []
    for i in range(len(cns)):
        ref = cns[:i] + cns[i+1:]
        scores.append(sentence_bleu([r.split() for r in ref], cns[i].split()))
    return np.mean(scores)

# Lexical Diversity
def lexical_diversity(cns):
    words = ' '.join(cns).split()
    return len(set(words)) / len(words)

# Perplexity
model_name = "ai4bharat/indic-bert"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
model.eval()

def calculate_perplexity(text):
    inputs = tokenizer(text, return_tensors="pt")
    input_ids = inputs.input_ids
    with torch.no_grad():
        outputs = model(input_ids, labels=input_ids)
        loss = outputs.loss
    return torch.exp(loss).item()

# Evaluation function
def evaluate_model(cns):
    return {
        "BLEU-1": compute_bleu(cns, 1),
        "BLEU-2": compute_bleu(cns, 2),
        "BLEU-3": compute_bleu(cns, 3),
        "ROUGE": rouge.compute(predictions=cns, references=texts),
        "BERTScore": bertscore.compute(predictions=cns, references=texts, lang="ta"),
        "Self-BLEU": compute_self_bleu(cns),
        "Lexical Diversity": lexical_diversity(cns),
        "Perplexity": np.mean([calculate_perplexity(cn) for cn in cns[:50]])
    }

# Run evaluation
print("\nEvaluation of Counter Narratives Dataset:")
print(evaluate_model(cns))
```
