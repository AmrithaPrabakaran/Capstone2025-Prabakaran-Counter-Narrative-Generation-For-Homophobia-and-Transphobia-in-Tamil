```python

!pip install bert_score
import pandas as pd
import matplotlib.pyplot as plt
from bert_score import score


gpt = pd.read_excel("/content/sample_data/Final_DHOW_Dataset_Tamil_CN_08_07.xlsx").iloc[50:150].copy()
gemini = pd.read_csv("/content/sample_data/Gemini_CN_extraction.csv")
deepseek = pd.read_csv("/content/sample_data/Deepseek_pairs.csv")
grok = pd.read_csv("/content/sample_data/Grok_hate_speech_pairs.csv")


def clean(df, model_name):
    df.columns = [c.strip().upper() for c in df.columns]
    df = df.rename(columns={
        "TEXT": "Text",
        "COUNTER NARRATIVES": "Counter_Narrative",
        "COUNTER_NARRATIVE": "Counter_Narrative"  # in case of different format
    })
    df["MODEL"] = model_name
    return df[["Text", "Counter_Narrative", "MODEL"]].dropna()

gpt = clean(gpt, "GPT-4o")
gemini = clean(gemini, "Gemini")
deepseek = clean(deepseek, "DeepSeek")
grok = clean(grok, "Grok")

# Combine all into one dataset
data = pd.concat([gpt, gemini, deepseek, grok], ignore_index=True)

# Step 1: Compute a simple complexity score (length of HS text)
data["HS_Complexity"] = data["Text"].apply(len)


median_complexity = data["HS_Complexity"].median()
complex_data = data[data["HS_Complexity"] >= median_complexity]


models = complex_data["MODEL"].unique()
bert_results = {}

for model in models:
    subset = complex_data[complex_data["MODEL"] == model]
    refs = subset["Text"].tolist()
    hyps = subset["Counter_Narrative"].tolist()

    P, R, F1 = score(hyps, refs, lang="ta", model_type="xlm-roberta-base", verbose=True)
    bert_results[model] = {
        "BERTScore_P": round(P.mean().item(), 4),
        "BERTScore_R": round(R.mean().item(), 4),
        "BERTScore_F1": round(F1.mean().item(), 4),
        "Avg_HS_Complexity": round(subset["HS_Complexity"].mean(), 2)
    }


bert_df = pd.DataFrame.from_dict(bert_results, orient="index")
print("\nBERTScore on Complex Hate Speech Examples:\n")
print(bert_df)

bert_df.to_csv("bert_score_complex_hs_only.csv")
print("\nSaved as 'bert_score_complex_hs_only.csv'")


bert_df[["BERTScore_F1"]].plot(kind="bar", legend=False, figsize=(8, 5))
plt.title("BERTScore-F1 on Complex Hate Speech (Tamil CNs, 100/model)")
plt.ylabel("F1 Score")
plt.ylim(0, 1)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# --------- INSTALL REQUIRED LIBRARIES ---------
!pip install nltk rouge-score bert-score transformers sentencepiece torch --quiet

# --------- IMPORT LIBRARIES ----------
import pandas as pd
import numpy as np
import nltk
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from nltk.translate.meteor_score import meteor_score
from rouge_score import rouge_scorer
from bert_score import score as bert_score
from transformers import AutoTokenizer, AutoModelForSequenceClassification, MBartForConditionalGeneration, MBart50TokenizerFast
import torch


nltk.download('punkt')
nltk.download('wordnet')


df = pd.read_excel("/content/sample_data/Final_DHOW_Dataset_Tamil_CN_08_07.xlsx")
df.columns = df.columns.str.strip().str.upper()
df = df.rename(columns={"TEXT": "Text", "COUNTER NARRATIVES": "Counter_Narrative"})
df = df[["Text", "Counter_Narrative"]].dropna().reset_index(drop=True)


smooth_fn = SmoothingFunction().method1
rouge = rouge_scorer.RougeScorer(['rouge1', 'rougeL'], use_stemmer=True)


refs = df["Text"].tolist()
hyps = df["Counter_Narrative"].tolist()
P, R, F1 = bert_score(hyps, refs, lang="ta", model_type="xlm-roberta-base", verbose=False)
bertscores = F1.tolist()


tox_tokenizer = AutoTokenizer.from_pretrained("unitary/toxic-bert")
tox_model = AutoModelForSequenceClassification.from_pretrained("unitary/toxic-bert")
tox_labels = ["toxic", "severe_toxic", "obscene", "threat", "insult", "identity_hate"]


mbart_model = MBartForConditionalGeneration.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
mbart_tokenizer = MBart50TokenizerFast.from_pretrained("facebook/mbart-large-50-many-to-many-mmt")
mbart_tokenizer.src_lang = "ta_IN"

def compute_perplexity(text):
    inputs = mbart_tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=128)
    input_ids = inputs["input_ids"]
    with torch.no_grad():
        outputs = mbart_model(input_ids=input_ids, labels=input_ids)
        loss = outputs.loss
    return torch.exp(loss).item()

def get_toxicity_scores(text):
    inputs = tox_tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512)
    with torch.no_grad():
        logits = tox_model(**inputs).logits
        probs = torch.sigmoid(logits).squeeze().numpy()
    return dict(zip(tox_labels, probs))

def distinct_n(corpus, n):
    ngrams = set()
    total = 0
    for sent in corpus:
        tokens = sent.split()
        total += max(0, len(tokens) - n + 1)
        for i in range(len(tokens) - n + 1):
            ngrams.add(tuple(tokens[i:i+n]))
    return len(ngrams) / total if total > 0 else 0


bleu1s, bleu2s, bleu3s, bleu4s = [], [], [], []
rouge1s, rougels, meteors, perplexities = [], [], [], []
toxicity_results = []

for i, row in df.iterrows():
    ref = [row["Text"].split()]
    hyp = row["Counter_Narrative"].split()

    bleu1s.append(sentence_bleu(ref, hyp, weights=(1, 0, 0, 0), smoothing_function=smooth_fn))
    bleu2s.append(sentence_bleu(ref, hyp, weights=(0.5, 0.5, 0, 0), smoothing_function=smooth_fn))
    bleu3s.append(sentence_bleu(ref, hyp, weights=(0.33, 0.33, 0.33, 0), smoothing_function=smooth_fn))
    bleu4s.append(sentence_bleu(ref, hyp, weights=(0.25, 0.25, 0.25, 0.25), smoothing_function=smooth_fn))

    r = rouge.score(" ".join(ref[0]), " ".join(hyp))
    rouge1s.append(r["rouge1"].fmeasure)
    rougels.append(r["rougeL"].fmeasure)

    meteors.append(meteor_score([ref[0]], hyp))
    perplexities.append(compute_perplexity(row["Counter_Narrative"]))
    toxicity_results.append(get_toxicity_scores(row["Counter_Narrative"]))


distinct1 = distinct_n(df["Counter_Narrative"].tolist(), 1)
distinct2 = distinct_n(df["Counter_Narrative"].tolist(), 2)


metrics_df = pd.DataFrame({
    "BLEU-1": bleu1s,
    "BLEU-2": bleu2s,
    "BLEU-3": bleu3s,
    "BLEU-4": bleu4s,
    "ROUGE-1": rouge1s,
    "ROUGE-L": rougels,
    "METEOR": meteors,
    "BERTScore-F1": bertscores,
    "Perplexity": perplexities
})

tox_df = pd.DataFrame(toxicity_results)
results_df = pd.concat([metrics_df, tox_df], axis=1)


summary = results_df.mean().to_dict()
summary["Distinct-1"] = distinct1
summary["Distinct-2"] = distinct2


results_df.to_csv("full_metrics_gpt4o_dataset.csv", index=False)
pd.DataFrame([summary]).to_csv("full_metrics_summary.csv", index=False)


print("Summary of Metrics:\n")
for k, v in summary.items():
    print(f"{k:15s}: {v:.4f}")
```

```python
import pandas as pd
import faiss
from sentence_transformers import SentenceTransformer


df = pd.read_csv("/content/sample_data/final_dataset.csv")
corpus = df['COUNTER NARRATIVES'].astype(str).tolist()


embed_model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')
embeddings = embed_model.encode(corpus, convert_to_numpy=True, show_progress_bar=True)

# Build FAISS index
index = faiss.IndexFlatL2(embeddings.shape[1])
index.add(embeddings)
```

```python
test_df = pd.read_csv("/content/sample_data/Final_test_Journal.csv")  # Must have column 'HS'
```

```python
def retrieve_cn(query, k=3):
    q_embed = embed_model.encode([query], convert_to_numpy=True)
    D, I = index.search(q_embed, k)
    return [corpus[i] for i in I[0]]

def build_debate_prompt(hs_text):
    retrieved = retrieve_cn(hs_text)
    prompt = f"Hate Speech: {hs_text}\n\nCounter-Narrative Agents Debate:\n"
    for i, cn in enumerate(retrieved):
        prompt += f"Agent {i+1}: {cn}\n"
    prompt += "\nFinal Counter-Narrative in Tamil:"
    return prompt
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

tokenizer = AutoTokenizer.from_pretrained("microsoft/phi-2")
model = AutoModelForCausalLM.from_pretrained("microsoft/phi-2")
model.eval()

def generate_cn(prompt):
    inputs = tokenizer(prompt, return_tensors="pt")
    with torch.no_grad():
        outputs = model.generate(**inputs, max_new_tokens=100)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
from tqdm import tqdm

results = []

for hs_text in tqdm(test_df['TEXT'], desc="Generating CNs"):
    prompt = build_debate_prompt(hs_text)
    cn_generated = generate_cn(prompt)
    results.append({'TEXT': hs_text, 'Generated_CN': cn_generated})

output_df = pd.DataFrame(results)
output_df.to_csv("generated_cn_output.csv", index=False)
```

```python
import pandas as pd
import faiss
from sentence_transformers import SentenceTransformer, util
import torch

# Load and encode corpus
df = pd.read_csv("/content/sample_data/final_dataset.csv")
corpus = df['COUNTER NARRATIVES'].astype(str).tolist()

embed_model = SentenceTransformer('paraphrase-multilingual-MiniLM-L12-v2')
corpus_embeddings = embed_model.encode(corpus, convert_to_numpy=True, show_progress_bar=True)

# Build FAISS index
index = faiss.IndexFlatL2(corpus_embeddings.shape[1])
index.add(corpus_embeddings)

# Function to retrieve + re-rank using FASIS-style alignment
def retrieve_cn_fasis(query, k=10, top_n=3):
    query_embed = embed_model.encode([query], convert_to_tensor=True)

    # Step 1: Retrieve top-k semantically similar from FAISS
    query_np = query_embed.cpu().detach().numpy()
    D, I = index.search(query_np, k)

    retrieved = [corpus[i] for i in I[0]]

    # Step 2: Re-rank based on semantic alignment (cosine sim between query and each CN)
    scores = []
    for cn in retrieved:
        cn_embed = embed_model.encode(cn, convert_to_tensor=True)
        similarity = util.pytorch_cos_sim(query_embed, cn_embed).item()
        scores.append((cn, similarity))

    # Sort by similarity (descending)
    sorted_cn = sorted(scores, key=lambda x: x[1], reverse=True)

    # Return top-n most factually/semantically aligned CNs
    return [cn for cn, score in sorted_cn[:top_n]]

# Build the debate prompt using FASIS-style CNs
def build_debate_prompt_fasis(hs_text):
    retrieved = retrieve_cn_fasis(hs_text)
    prompt = f"Hate Speech: {hs_text}\n\nCounter-Narrative Agents Debate:\n"
    for i, cn in enumerate(retrieved):
        prompt += f"Agent {i+1}: {cn}\n"
    prompt += "\nFinal Counter-Narrative in Tamil:"
    return prompt
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM

tokenizer = AutoTokenizer.from_pretrained("microsoft/phi-2")
model = AutoModelForCausalLM.from_pretrained("microsoft/phi-2")
model.eval()

def generate_cn(prompt):
    inputs = tokenizer(prompt, return_tensors="pt")
    with torch.no_grad():
        outputs = model.generate(**inputs, max_new_tokens=100)
    return tokenizer.decode(outputs[0], skip_special_tokens=True)
```

```python
import pandas as pd
import nltk
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from rouge_score import rouge_scorer
import evaluate
from detoxify import Detoxify

# Download required nltk data
nltk.download('punkt', force=True)

# Load the reference (seed) and generated data
reference_file = "/content/sample_data/Seed_dataset.xlsx"  # Seed data with ground-truth CNs
prediction_file = "/content/sample_data/vivatham_agents_final_cn.xlsx"  # Model-generated CNs

# Read both files
ref_df = pd.read_excel(reference_file)
pred_df = pd.read_excel(prediction_file)

# Filter to match only valid TEXTs
valid_texts = pred_df['TEXT'].unique()
ref_df = ref_df[ref_df['TEXT'].isin(valid_texts)]

# Merge on the 'TEXT' column
merged_df = pd.merge(ref_df[['TEXT', 'COUNTER NARRATIVES']], pred_df[['TEXT', 'Final_CN']], on='TEXT')

# Extract references and predictions
references = merged_df['COUNTER NARRATIVES'].fillna("").astype(str).tolist()
predictions = merged_df['Final_CN'].fillna("").astype(str).tolist()

# Initialize scorers
smoother = SmoothingFunction().method4
rouge = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
bertscore = evaluate.load("bertscore")

# Compute BLEU scores
bleu1 = [sentence_bleu([ref.split()], pred.split(), weights=(1, 0, 0, 0), smoothing_function=smoother) for ref, pred in zip(references, predictions)]
bleu2 = [sentence_bleu([ref.split()], pred.split(), weights=(0.5, 0.5, 0, 0), smoothing_function=smoother) for ref, pred in zip(references, predictions)]
bleu3 = [sentence_bleu([ref.split()], pred.split(), weights=(0.33, 0.33, 0.33, 0), smoothing_function=smoother) for ref, pred in zip(references, predictions)]
bleu4 = [sentence_bleu([ref.split()], pred.split(), weights=(0.25, 0.25, 0.25, 0.25), smoothing_function=smoother) for ref, pred in zip(references, predictions)]

# ROUGE
rouge_scores = [rouge.score(pred, ref) for pred, ref in zip(predictions, references)]

# BERTScore (Tamil)
bertscore_result = bertscore.compute(predictions=predictions, references=references, lang="ta")

# Distinct n-grams
unigrams = set()
bigrams = set()
for t in predictions:
    tokens = nltk.word_tokenize(t)
    unigrams.update(tokens)
    bigrams.update(nltk.bigrams(tokens))
distinct1 = len(unigrams)
distinct2 = len(bigrams)

# Novelty
novelty_score = sum([1 - (len(set(p.split()) & set(r.split())) / len(set(p.split()))) if len(set(p.split())) > 0 else 0 for p, r in zip(predictions, references)]) / len(predictions)

# Toxicity
toxicity = Detoxify("original").predict(predictions)
non_toxicity = [1 - x for x in toxicity['toxicity']]

# Composite Score (one version)
composite_score = (
    0.3 * sum(bertscore_result['f1']) / len(bertscore_result['f1']) +
    0.2 * (sum(non_toxicity) / len(non_toxicity)) +
    0.15 * sum(bleu4) / len(bleu4) +
    0.15 * sum([r['rougeL'].fmeasure for r in rouge_scores]) / len(rouge_scores) +
    0.2 * novelty_score
)

# Print summary
print("\nSummary of Evaluation Metrics:\n")
print(f"BLEU-1: {sum(bleu1)/len(bleu1):.4f}")
print(f"BLEU-2: {sum(bleu2)/len(bleu2):.4f}")
print(f"BLEU-3: {sum(bleu3)/len(bleu3):.4f}")
print(f"BLEU-4: {sum(bleu4)/len(bleu4):.4f}")
print(f"ROUGE-L: {sum([r['rougeL'].fmeasure for r in rouge_scores])/len(rouge_scores):.4f}")
print(f"BERTScore: {sum(bertscore_result['f1'])/len(bertscore_result['f1']):.4f}")
print(f"Novelty: {novelty_score:.4f}")
print(f"Distinct-1: {distinct1:.4f}")
print(f"Distinct-2: {distinct2:.4f}")
print(f"Toxicity: {sum(toxicity['toxicity']) / len(toxicity['toxicity']):.4f}")
print(f"Severe Toxicity: {sum(toxicity['severe_toxicity']) / len(toxicity['severe_toxicity']):.4f}")
print(f"Obscene: {sum(toxicity['obscene']) / len(toxicity['obscene']):.4f}")
print(f"Identity Attack: {sum(toxicity['identity_attack']) / len(toxicity['identity_attack']):.4f}")
print(f"Insult: {sum(toxicity['insult']) / len(toxicity['insult']):.4f}")
print(f"Threat: {sum(toxicity['threat']) / len(toxicity['threat']):.4f}")
print(f"Non-Abusive: {sum(non_toxicity)/len(non_toxicity):.4f}")
print(f"Composite Score: {composite_score:.4f}")
```
